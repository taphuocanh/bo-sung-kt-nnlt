```
import pandas as pd
import matplotlib.pyplot as plt

# Đọc dữ liệu từ file csv
data = pd.read_csv('/content/drive/My Drive/Book1.csv')
largest_units_items = data.sort_values(by='Units', ascending=False).head(5)
print(largest_units_items)

units = largest_units_items.loc[:,"Units"]
items = largest_units_items.loc[:,"Item"]

plt.style.use('ggplot')
x_pos = [i for i, _ in enumerate(items)]
plt.bar(x_pos, units, color='green')
plt.xlabel("Item")
plt.ylabel("Units")
plt.title("Biểu đồ cột dọc")

plt.xticks(x_pos, items)

plt.show()

plt.barh(x_pos, units, color='Blue')
plt.ylabel("Item")
plt.xlabel("Units")
plt.title("Biểu đồ cột ngang")
plt.yticks(x_pos, items)

plt.show()


x=range(len(items))
plt.xticks(x,items)
plt.plot(x, units)
plt.ylim(ymin=units.min() - 5)  # this line
plt.xlabel("Item")
plt.ylabel("Units")
plt.title("Biểu đồ đường")
plt.show()

plt.boxplot([data.loc[:,"Units"].sort_values()], labels=['Units'])
plt.title("Biểu đồ box")
plt.show()
```
